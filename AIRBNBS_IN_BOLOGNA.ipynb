```python
# looking at the price column

tt['price'] = pd.to_numeric(tt['price'].str.replace('$', '').str.replace(',', ''), errors='coerce')
tt['price']
```

```python
# identifying outliers


q1 = tt['price'].quantile(0.25)
q3 = tt['price'].quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = tt[(tt['price'] < lower_bound) | (tt['price'] > upper_bound)]

print(f"Number of outliers: {len(outliers)}")
```

```python
# removing outliers

tt_clean = tt[(tt['price'] >= lower_bound) & (tt['price']<= upper_bound)]
print(f"No. of rows remaining: {len(tt_clean)}")
```

```python
print(tt_clean['price'].describe())
```

```python
sns.boxplot(x=tt_clean['price'])
plt.title('Price Distribution After Handling Outliers')
plt.show()
```

```python
# removing outliers for the other columns

cols = ['bedrooms', 'bathrooms', 'beds']

for column in cols:
    q1 = tt_clean[column].quantile(0.25)
    q3 = tt_clean[column].quantile(0.75)
    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr


    outliers = tt_clean[(tt_clean[column] < lower_bound) | (tt_clean[column] > upper_bound)]

    print(f"Number of outliers in {column}: {len(outliers)}")
```

```python
tt_clean = tt_clean[(tt_clean['bedrooms'] >= lower_bound) & (tt_clean['bedrooms'] <= upper_bound)]
tt_clean = tt_clean[(tt_clean['bathrooms'] >= lower_bound) & (tt_clean['bathrooms'] <= upper_bound)]
tt_clean = tt_clean[(tt_clean['beds'] >= lower_bound) & (tt_clean['beds'] <= upper_bound)]
```
